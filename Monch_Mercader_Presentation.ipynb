```python
#Importing all the libraries and data visualization tools

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
plt.rcParams['figure.figsize'] = [6.0, 6.0] # control the width and height of all figures
plt.rcParams['figure.dpi'] = 500 # control the quality of all figures. Higher is better :)

%matplotlib inline
```

```python
#Loading in the data sets, these were exports from Google Trends using the search key: "diet"
#Loading the interest over time (5 years, Dec 2015 to June 2020) 
raw_diet = pd.read_csv('../data/Google_search/Diet_Google_2015_2020.csv', sep=',', index_col = 'Week', skiprows=1)
```

## *CSV data source = https://trends.google.com/trends/explore?q=diet&geo=GB

```python
raw_diet['diet: (Australia)'].plot(kind='line', use_index=True, figsize=(30, 8))
plt.title('Searching Diet from 2015 to 2020', fontsize=20, color='Blue')
plt.xlabel('Weeks', fontsize=20)
plt.ylabel('Popularity 0-100%', fontsize=20)
plt.show()
```

### Represent growth of search in percentage over time, here it is visualized over a 5 year period.

```python
#Trying to see the search popularity by region

diet_region = pd.read_csv('../data/Google_search/Diet_by_Region.csv', sep=',', header=0, skiprows=2)
```

```python
fig, ax01 = plt.subplots(1, 1)

# Example data
x_pos01 = np.arange(len(diet_region['Region']))
popularity01 = diet_region['diet: (12/11/15 - 12/11/20)']

ax01.set_title('Diet search by Region')
ax01.bar(x_pos01, popularity01, width = 0.8, align='center', color='Red')
plt.ylabel('Popularity by 0-100%')
plt.xlabel('Region')
plt.xticks(x_pos01, diet_region['Region'], rotation = 90)
plt.show()
```

```python
# Representing searches based on Region

fig, ax = plt.subplots(1, 1)

# Example data
y_pos = np.arange(len(diet_region['Region']))
popularity = diet_region['diet: (12/11/15 - 12/11/20)']

ax.set_title('Diet search by Regions')
ax.barh(y_pos, popularity, align='center', height=0.75, color='#e34f4f')
ax.set_yticks(y_pos)
ax.set_yticklabels(diet_region['Region'], fontsize=12)
ax.invert_yaxis()
ax.set_xlabel('Popularity 0-100%')

plt.show()
```

### Using (Horizontal) Bars are good in showing side-by-side values across items in the same category. Here it shows the growth of the searches over the region within the timeframe
